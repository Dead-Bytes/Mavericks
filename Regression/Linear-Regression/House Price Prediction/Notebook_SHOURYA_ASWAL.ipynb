```python
import numpy as np
import pandas as pd
dataset = pd.read_csv('data.csv')
#dataset=dataset[dataset['price']<1000000]
#dataset.info()
target=dataset[['price']]
features=dataset[['sqft_area','bedrooms','bathrooms']]
from sklearn.model_selection import train_test_split

fTrain, fTest, tTrain, tTest = train_test_split(features,target,test_size=0.2)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(fTrain,tTrain)

print(model.score(fTest,tTest))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('data.csv')
#dataset=dataset[dataset['price']<1000000]  #removing outliers
y=dataset.iloc[:,0]
x=dataset.iloc[:,3]
X=[]
Y=[]
Z=[]
W=range(0,800)

from scipy import stats
for i in x:
  X.append(float(i/10))
for i in y:
  Y.append(float(i/1000))
slope, intercept,correlation_coefficient,p_value,stand_dev = stats.linregress(X,Y)
for i in range(0,800):
  Z.append(intercept+slope*i)

plt.plot(X,Y,'go')
plt.plot(W,Z,'b')
plt.axis((0,800,0,1000))
print(correlation_coefficient,'\n',correlation_coefficient**2,'\n', p_value)
```
